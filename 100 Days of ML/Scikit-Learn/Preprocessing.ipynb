```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('drawndata1.csv')
```

```python
df.head(3)
```

```python
X = df[['x','y']].values
y = df['z'] == 'a'
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
```

```python
# using standard scale
from sklearn.preprocessing import StandardScaler, QuantileTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline 
# X_new = StandardScaler().fit_transform(X)

X_new = QuantileTransformer(n_quantiles=100).fit_transform(X) #using qt for scaling the data in a better manner
# n_quantiles = 100 bcs by default it is set to 1000 and we do not have that much data

plt.scatter(X_new[:,0],X_new[:,1],c=y)
```

```python
# will make a function named plot_output to see the effect of scaling on 
# predicted data
def plot_output(scaler):
    pipe = Pipeline([
        ("scale", scaler),
        ("model", KNeighborsClassifier(n_neighbors=20, weights='distance'))
    ])

    pred = pipe.fit(X, y).predict(X)

    plt.figure(figsize=(9, 3))
    plt.subplot(131)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.title("Original Data")
    plt.subplot(132)
    X_tfm = scaler.transform(X)
    plt.scatter(X_tfm[:, 0], X_tfm[:, 1], c=y)
    plt.title("Transformed Data")
    plt.subplot(133)
    X_new = np.concatenate([
        np.random.uniform(0, X[:, 0].max(), (5000, 1)), 
        np.random.uniform(0, X[:, 1].max(), (5000, 1))
    ], axis=1)
    y_proba = pipe.predict_proba(X_new)
    plt.scatter(X_new[:, 0], X_new[:, 1], c=y_proba[:, 1], alpha=0.7)
    plt.title("Predicted Data")
```

```python
plot_output(scaler = StandardScaler())
```

```python
plot_output(scaler=QuantileTransformer(n_quantiles=100))
```
