```python
import numpy as np
import pandas as pd
from sklearn import neighbors, metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('car.data')
```

```python
data
```

```python
# we will only consider buying, maintenance,safety 
X = data[['buying', 'maintenance','safety']].values
y = data[['class']]
print(X,y)
```

```python
# converting the string data to int
Le = LabelEncoder()
for i in range(len(X[0])):
    X[:, i] = Le.fit_transform(X[:, i])

print(X)
```

```python
label_mapping = {
    'unacc':0,
    'acc':1,
    'good':2,
    'vgood':3
}

y['class'] = y['class'].map(label_mapping)
y = np.array(y)
print(y)
```

```python
# create a model
knn = neighbors.KNeighborsClassifier(n_neighbors=25, weights='uniform')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.2,random_state=69)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
acc = metrics.accuracy_score(y_test,y_pred)
```

```python
print("The predictions are: ",y_pred)
print("The accuracy score is ",acc)
```
